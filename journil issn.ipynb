```python
from collections import defaultdict
from crossref.restful import Journals
import csv
import json
lista=[]
listag=[]
columns= defaultdict(list)
j = Journals()
with open('sbcf1.csv', newline='') as csvfile:
 spamreader = csv.reader(csvfile)
 for row in spamreader:
        lista.append((row))
for row in lista:
  n=",".join(row)
  m=n.strip(";")
  #print(m)
  x=j.journal(m)
  #print(x)
  listag.append(x)
print (listag)
with open('1036-fin.json', 'w') as json_file:
     json.dump(listag, json_file)
```
